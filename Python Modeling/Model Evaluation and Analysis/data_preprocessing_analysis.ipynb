```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns',50)
```

```python
df1 = pd.read_csv('partially_cleaned_survey.csv')
df1.head(3)
```

```python
# Add your code here. Feel free to add more code cells or markdown cells 
# as you see fit (and this advice applies to all later questions as well).
df1.loc[df1['BachTime'] == '< 1 year ago', 'BachTime'] = 'Bach_0to1'
df1.loc[df1['BachTime'] == 'longer than 1 year ago but less than 3 years ago', 'BachTime'] = 'Bach_1to3'
df1.loc[df1['BachTime'] == 'longer than 3 years ago but less than 5 years ago', 'BachTime'] = 'Bach_3to5'
df1.loc[df1['BachTime'] == 'over 5 years ago', 'BachTime'] = 'Bach_5plus'
print(df1)
```

```python
def bachtime(bachtime):
    if bachtime == '< 1 year ago':
        return 'Bach_0to1'
    elif bachtime == 'longer than 1 year ago but less than 3 years ago':
        return 'Bach_1to3'
    elif bachtime == 'longer than 3 years ago but less than 5 years ago':
        return 'Bach_3to5'
    elif bachtime == 'over 5 years ago':
        return 'Bach_5plus'
    else:
        return bachtime
df1['BachTime2'] = df1['BachTime'].apply(bachtime)
print(df1)
```

```python
df1['BachTime3'] = df1['BachTime'].replace({'< 1 year ago': 'Bach_0to1',
                                             'longer than 1 year ago but less than 3 years ago': 'Bach_1to3',
                                             'longer than 3 years ago but less than 5 years ago': 'Bach_3to5',
                                             'over 5 years ago': 'Bach_5plus'})
print(df1)
```

```python
javasql = ((df1['Java'] == 1) & (df1['SQL'] == 1)).value_counts()[1]
javasqlperc = (javasql / (df1['Java'] == 1).sum()) * 100
print(javasqlperc)

javaexcel = ((df1['Java'] == 1) & (df1['Excel'] == 1)).value_counts()[1]
javaexcelperc = (javaexcel / (df1['Java'] == 1).sum()) * 100
print(javaexcelperc)

nojavasql = ((df1['Java'] == 0) & (df1['SQL'] == 1)).value_counts()[1]
nojavasqlperc = (nojavasql / (df1['Java'] == 0).sum()) * 100
print(nojavasqlperc)

nojavaexcel = ((df1['Java'] == 0) & (df1['Excel'] == 1)).value_counts()[1]
nojavaexcelperc = (nojavaexcel / (df1['Java'] == 0).sum()) * 100
print(nojavaexcelperc)
```
